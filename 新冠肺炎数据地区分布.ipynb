```python
import numpy as np
import pandas as pd

data = pd.read_csv("肺炎清洗数据.csv")
data = data.fillna(0)
data
```

```python
cov_data = pd.pivot_table(data,index=['地区'],values=['新增确诊'],aggfunc='count')
print(cov_data)
```

```python
from pyecharts import Map

list_data = dict(zip(list(cov_data.index), list(cov_data['新增确诊'])))
list_data
```

```python
province = list(list_data.keys())   #将字典中的省份key以列表的形式取出来
values = list(list_data.values())

map = Map("疫情地图-中国累计确诊", width=1200, height=800,title_pos="left",title_color="#404a59")
map.add("累计确诊",
#输入数据
province,
values,
#设置分类颜色
pieces = [
    {"max": 0, "label": "0人", "color": "#3CB371"},
    {"min": 1, "max": 9, "label": "1-10人", "color": "#83AF9B"},
    {"min": 10, "max": 99, "label": "10-99人", "color": "#C8C8A9"},
    {"min": 100, "max": 249, "label": "100-499人", "color": "#F9CDAD"},
    {"min": 250, "max": 499, "label": "500-999人", "color": "#F08080"},
    {"min": 500, "max": 999, "label": "500-10000人", "color": "#CD5C5C"},
    {'min': 1000, "label": ">1000人", "color": "#FE4365"}],  # 不指定 max，表示 max 为无限大
isual_range = [0, 1000],
is_piecewise = True,#将分类颜色分段，利于分辨
is_visualmap = True,#将数据映射至地图上
visual_split_number = 5,#可视化数组分组
maptype = 'china',#设置地图类型，需要提前下载地图包
visual_text_color = '#000',#标签颜色
is_label_show = True,# 显示各区域名称
is_map_symbol_show = False)
map.render('疫情地图-全国累计确诊.html')#生成map
map
```

```python
area_data = pd.read_csv("最终数据（副本）.csv")
area_data['情感指数'] = area_data['积极数据']-area_data['消极数据']
area_data['点赞数'] = area_data['点赞数']+1
area_data['情感得分'] = area_data['情感指数']*area_data['点赞数']
area_data
```

```python
motion_data = pd.pivot_table(area_data,index=['地区'],values=['情感得分'],aggfunc='sum')
print(motion_data)
```

```python
motion_list_data = dict(zip(list(motion_data.index), list(motion_data['情感得分'])))
motion_list_data
```

```python
motion_province = list(motion_list_data.keys())   #将字典中的省份key以列表的形式取出来
motion_values = list(motion_list_data.values())

map = Map("情感地图-全国累计微博情感", width=1200, height=800,title_pos="left",title_color="#404a59")
map.add("情感累计",
#输入数据
motion_province,
motion_values,
#设置分类颜色
pieces = [
    {"max": -1000, "label": "特别消极", "color": "#FE4365"},
    {"min": -999, "max": -500,"label": "消极", "color": "#CD5C5C"},
    {"min": -499, "max": -100, "label": "比较消极", "color": "#F08080"},
    {"min": -99, "max": 99, "label": "平淡", "color": "#F9CDAD"},
    {"min": 100, "max": 499,"label": "比较积极", "color": "#C8C8A9"},
    {"min": 500, "max": 999, "label": "积极", "color": "#83AF9B"},
    {'min': 1000, "label": "特别积极", "color": "#3CB371"}],  # 不指定 max，表示 max 为无限大
isual_range = [0, 1000],
is_piecewise = True,#将分类颜色分段，利于分辨
is_visualmap = True,#将数据映射至地图上
visual_split_number = 5,#可视化数组分组
maptype = 'china',#设置地图类型，需要提前下载地图包
visual_text_color = '#000',#标签颜色
is_label_show = True,# 显示各区域名称
is_map_symbol_show = False)
map.render('情感地图-全国累计微博情感.html')#生成map
map
```
