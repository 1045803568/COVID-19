```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('评论者信息(1).csv',error_bad_lines=False)
df1
```

```python
df2 = pd.read_csv('评论者信息(2).csv',error_bad_lines=False)
df2
```

```python
df1['评论'] = df1.apply(lambda row : row['人名_评论'].replace(str(row['人名']+'：'),''),axis=1)
df1['评论'] = df1.apply(lambda row : row['评论'].replace('\n ',''),axis=1)
df1['点赞数'] = df1.apply(lambda row : row['点赞数'].replace('赞','0'),axis=1)
df1
```

```python
df2['评论'] = df2.apply(lambda row : row['人名_评论'].replace(str(row['人名']+'：'),''),axis=1)
df2['评论'] = df2.apply(lambda row : row['评论'].replace('\n ',''),axis=1)
df2['点赞数'] = df2.apply(lambda row : row['点赞数'].replace('赞','0'),axis=1)
df2
```

```python
df3 = pd.concat([df1,df2])
df3 = df3.reset_index(drop = True)
df3 = df3.drop(['人名_评论','用户主页网址'],axis=1)
df3
```

```python
df4 = pd.read_csv('用户地区信息前13000条.csv',error_bad_lines=False)
df4['地区1'] = df4.apply(lambda row : str(row['文本1']).replace('\r','').replace('\n','').replace('\t',''),axis=1)
df4['地区2'] = df4.apply(lambda row : str(row['文本2']).replace('\r','').replace('\n','').replace('\t','').replace('2 ',''),axis=1)
df4 = df4.drop(['文本1','文本2'],axis=1)
df4
```

```python
df5 = pd.read_csv('用户信息后8000条.csv',error_bad_lines=False)
df5['地区1'] = df5.apply(lambda row : str(row['文本1']).replace('\r','').replace('\n','').replace('\t','').replace('2 ',''),axis=1)
df5['地区2'] = df5.apply(lambda row : str(row['文本2']).replace('\r','').replace('\n','').replace('\t','').replace('2 ',''),axis=1)
df5 = df5.drop(['文本1','文本2'],axis=1)
df5
```

```python
df6 = pd.concat([df4,df5])
df6 = df6.reset_index(drop = True)
df6.to_csv('用户地区数据.csv') 
```

```python
df7 = pd.read_csv('用户地区数据（清洗后）.csv',error_bad_lines=False)
df7 = df7.drop(['Unnamed: 0'],axis=1)
df7
```
