```python
import pandas as pd
import numpy as np

```

```python

df = pd.read_csv('rmrb.csv',delimiter='\t',error_bad_lines=False)
df = df.drop(['原始图片url','发布位置','发布工具'],axis=1)
df
```

```python
b1 = []
for i in range(len(df)):
    a = "本土确诊病例"
    #这里的关键词是随便选取的
    if a in df.loc[i, '微博正文']:
        a1 = df.loc[i, '微博id']
        b1.append(a1)

f1 = pd.DataFrame(columns=['微博id'])

f1['微博id'] = b1
f1["URL"] = 'https://weibo.com/2803301701/'+ f1["微博id"]
print(f1)
```
